```python
import pandas as pd
import numpy as np 
import collections
import googlemaps
import matplotlib.pyplot as plt
```

```python
buses_data = pd.read_csv('Potentail_Bust_Stops.csv')
employee_data = pd.read_csv('Employee_Addresses.csv')
```

```python
print(buses_data.shape)
buses_data.head()
```

```python
print(employee_data.shape)
employee_data.head()
```

```python
def parseStreet(x):
    l = x.split(',')
    l = l[0].split(' ')
    streetName = ' '.join(l[1:])
    streetName = streetName.upper()
    return streetName
```

```python
employee_data['street'] = employee_data['address']
```

```python
employee_data['street'] = employee_data['street'].map(parseStreet)
```

```python
street_table = collections.Counter(employee_data['street'])
```

```python
buses_data['employeeNum'] = 0 
```

```python
lookup = lambda x: street_table.get(x, 0)
buses_data['employeeNum'] = buses_data['Street_One'].map(lookup) + buses_data['Street_Two'].map(lookup)
```

```python
buses_data.sort_values('employeeNum', inplace =True, )
```

```python
employee_data['lat'] = 0 
employee_data['lng'] = 0
buses_data['address'] = buses_data['Street_One'] + ' and ' + buses_data['Street_Two'] + ',San Francisco, CA 94112, USA'
buses_data['lat'] = 0 
buses_data['lng'] = 0 
```

```python
def geoCoding(x):
    geocode_result = gmaps.geocode(x)
    location = geocode_result[0]['geometry']['location']
    return [location['lat'], location['lng']]
```

```python
 employee_data['location'] =employee_data['address'].map(geoCoding)
```

```python
 buses_data['location'] = buses_data['address'].map(geoCoding)
```

```python
buses_data['lat'] = buses_data['location'].map(lambda x: x[0])
buses_data['lng'] = buses_data['location'].map(lambda x: x[1])
```

```python
employee_data['lat'] = employee_data['location'].map(lambda x: x[0])
employee_data['lng'] = employee_data['location'].map(lambda x: x[1])
```

```python
from sklearn.cluster import KMeans
X = employee_data[['lat', 'lng']]
kmeans = KMeans(n_clusters=14, random_state=0).fit(X)
```

```python
%matplotlib notebook
plt.scatter(employee_data['lat'], employee_data['lng'], c=kmeans.labels_.astype(np.float) )
plt.scatter([element[0] for element in kmeans.cluster_centers_], [element[1] for element in kmeans.cluster_centers_], color = 'red')
```

```python
%matplotlib notebook
plt.scatter(buses_data['lat'], buses_data['lng'])
plt.scatter([element[0] for element in kmeans.cluster_centers_], [element[1] for element in kmeans.cluster_centers_], color = 'red')
```
